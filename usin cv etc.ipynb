Import needed libraries

```python
import cv2 as cv
import imutils as im
import os
import glob
```

```python
os.chdir("frames1_thresh")
```

Get all image files.

```python
file = glob.glob("*.jpg")
```

Instantiate HOG descriptor and run all images through model.

```python
hog = cv.HOGDescriptor()
hog.setSVMDetector(cv.HOGDescriptor_getDefaultPeopleDetector())
```

```python
nums = []

for im in file:
    x = cv.imread(im, cv.IMREAD_GRAYSCALE)
    (regions, _) = hog.detectMultiScale(x, winStride=(1, 1), padding=(1, 1), scale=1.05)
    nums.append(len(regions))
```

Print all numbers of people predicted by model.

```python
print(nums)
```

Use pandas to create csv file for use when evaluating model later.

```python
import pandas as pd
```

```python
df = pd.DataFrame(nums)
```

```python
df.to_csv("CVTest.csv")
```
