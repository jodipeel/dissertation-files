```python
import cv2 as cv
import glob
import PIL as pil
from PIL import ImageFilter
import os
```

```python
os.chdir("frames1")
```

```python
imgs = glob.glob("*.jpg")
print(imgs)
```

```python
for img in imgs:
    x = cv.imread(img, cv.IMREAD_GRAYSCALE)
    _, thresh = cv.threshold(x, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
    cv.imwrite("thresh_" + img, thresh)
```
