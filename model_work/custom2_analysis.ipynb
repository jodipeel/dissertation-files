```python
real = real['count']
```

```python
real = real.truncate(after=199)
```

```python
def calc_acc(list):
    acc = []
    for i in range(len(list)):
        x = list[i].item()/real[i].item()
        acc.append(x)
    return acc
```

```python
YOLO2_acc = calc_acc(custYOLO2)
```
